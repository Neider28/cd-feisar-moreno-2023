```python
import random
import matplotlib.pyplot as plt

puntos = 1000000
p_lado = int(puntos**0.5)
separacion = 1 / (p_lado - 1)

# Genera coordenadas de puntos uniformemente distribuidos
coordenadas_x = []
coordenadas_y = []
puntos_dentro_x = []
puntos_dentro_y = []
puntos_fuera_x = []
puntos_fuera_y = []

for i in range(p_lado):
    x = i * separacion
    for j in range(p_lado):
        y = j * separacion
        coordenadas_x.append(x)
        coordenadas_y.append(y)

        # Verifica si el punto est√° dentro de la 1/4 parte de la circunferencia
        if x ** 2 + y ** 2 <= 1:
            puntos_dentro_x.append(x)
            puntos_dentro_y.append(y)
        else:
            puntos_fuera_x.append(x)
            puntos_fuera_y.append(y)


# Grafica los puntos dentro de la 1/4 parte de la circunferencia en azul
plt.scatter(puntos_dentro_x, puntos_dentro_y, s=0.5, color='blue')

# Grafica los puntos fuera de la 1/4 parte de la circunferencia en rojo
plt.scatter(puntos_fuera_x, puntos_fuera_y, s=0.5, color='red')

# Dibuja la circunferencia
theta = [i * math.pi / 2 / p_lado for i in range(p_lado)]
circunferencia_x = [math.cos(t) for t in theta]
circunferencia_y = [math.sin(t) for t in theta]
plt.plot(circunferencia_x, circunferencia_y, color='black')

plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos Distribuidos uniformemente')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import math

puntos = 1000000
p_lado = int(puntos ** 0.5)

# Genera coordenadas de puntos uniformemente distribuidos
coordenadas_x = []
coordenadas_y = []
puntos_dentro_x = []
puntos_dentro_y = []
puntos_fuera_x = []
puntos_fuera_y = []

for _ in range(puntos):
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    coordenadas_x.append(x)
    coordenadas_y.append(y)

    # Verifica si el punto est√° dentro de la 1/4 parte de la circunferencia
    if x ** 2 + y ** 2 <= 1:
        puntos_dentro_x.append(x)
        puntos_dentro_y.append(y)

    else:
        puntos_fuera_x.append(x)
        puntos_fuera_y.append(y)

# Grafica los puntos dentro de la 1/4 parte de la circunferencia de manera aleatoria en azul
plt.scatter(puntos_dentro_x, puntos_dentro_y, s=0.5, color='blue')

# Grafica los puntos fuera de la 1/4 parte de la circunferencia en rojo
plt.scatter(puntos_fuera_x, puntos_fuera_y, s=0.5, color='red')

# Dibuja la circunferencia
theta = [i * math.pi / 2 / p_lado for i in range(p_lado)]
circunferencia_x = [math.cos(t) for t in theta]
circunferencia_y = [math.sin(t) for t in theta]
plt.plot(circunferencia_x, circunferencia_y, color='black')

plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos Aleatorios')
plt.show()
```
