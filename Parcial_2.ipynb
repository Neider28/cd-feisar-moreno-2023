```python
def matriz_resolucion (min_X, max_X, min_Y, max_Y, resolucion):
  matriz_cols_X = math.ceil(((max_X-min_X)/resolucion))
  matriz_rows_Y = math.ceil(((max_Y-min_Y)/resolucion))

  return matriz_cols_X, matriz_rows_Y
```

```python
resolucion = 10
cols_X, rows_Y = matriz_resolucion(df_coordenadas['x'].min(), df_coordenadas['x'].max(), df_coordenadas['y'].min(), df_coordenadas['y'].max(), resolucion)
print(cols_X, rows_Y)
```

```python
matriz_ceros = np.zeros((rows_Y, cols_X))
matriz_ceros
```

```python
for index, fila in df_coordenadas.iterrows():
      X = fila['x'] # Obtener el valor de X de la fila correspondiente
      Y = fila['y'] # Obtener el valor de Y de la columna correspondiente
      resultado_X = math.floor((X - df_coordenadas['x'].min()) / resolucion)
      resultado_Y = math.floor((Y - df_coordenadas['y'].min()) / resolucion)
      matriz_ceros [resultado_Y, resultado_X]+=1

print(matriz_ceros)
```

```python
def distancia(x1,y1,x2,y2):
  return math.sqrt((x2-x1)**2 + (y2-y1)**2)
```

```python
radio_circulo = 100
n_resolucion = math.ceil(radio_circulo/resolucion)
```

```python
matriz_densidad = np.zeros((rows_Y, cols_X))
```

```python
def kernel(tipo_f, distancia, radio):
  if (tipo_f == "uniforme"):
    resultado = 1
  elif (tipo_f == "triangular"):
    resultado = 1-(distancia/radio)
  elif (tipo_f == "gauss"):
    resultado = math.exp(-(distancia**2/(2*radio**2)))
  elif (tipo_f == "epane"):
    resultado = 1-(distancia/radio)**2
  elif (tipo_f == "exponencial"):
    resultado = math.exp(-(distancia/radio))
  elif (tipo_f == "coseno"):
    resultado = math.cos((math.pi*distancia)/(2*radio))
  return resultado
```

```python
#las funciones kernel son: uniforme, triangular, gauss, epane, exponencial y coseno
kernel("exponencial",85,100)
```

```python
for i in range(rows_Y):
    Y_inicial, Y_final = np.clip([i - n_resolucion, i + n_resolucion], 0, rows_Y - 1)
    for j in range(cols_X):
        X_inicial, X_final = np.clip([j - n_resolucion, j + n_resolucion], 0, cols_X - 1)
        suma = 0
        for k in range(Y_inicial, Y_final + 1):
            for l in range(X_inicial, X_final + 1):
                d_centro = distancia(i, j, k, l)
                if (d_centro <= n_resolucion):
                    suma += matriz_ceros[k][l]*(kernel("exponencial",(d_centro*resolucion),radio_circulo))

        matriz_densidad[i][j] = suma/(math.pi*radio_circulo**2)
```

```python
import colorsys

def rgb_a_hsv(r, g, b):
  (h, s, v) = colorsys.rgb_to_hsv(r, g, b)
  return (h, s, v)

def hsv_a_rgb(h, s, v):
  (r, g, b) = colorsys.hsv_to_rgb(h, s, v)
  r = int(r)
  g = int(g)
  b = int(b)
  return (r, g, b)
```

```python
#rgb

def pintar_color(rojo,verde,azul):
  img  = Image.new('RGB', (300, 25))
  draw = ImageDraw.Draw(img)
  draw.rectangle([(0,0), img.size], fill = (rojo,verde,azul))
  display(img)


#Color inicial (red,green,blue)
r_inicial = 0
g_inicial = 255
b_inicial = 0
hsv_inicial = rgb_a_hsv(r_inicial, g_inicial, b_inicial)

#color final (red,green,blue)

r_final = 255
g_final = 0
b_final = 0
hsv_final = rgb_a_hsv(r_final, g_final, b_final)

n_colores = 10

for i in range(n_colores):
  h = (hsv_inicial[0]+i*((hsv_final[0]-hsv_inicial[0]) / (n_colores-1)))
  s = (hsv_inicial[1]+i*((hsv_final[1]-hsv_inicial[1]) / (n_colores-1)))
  v = (hsv_inicial[2]+i*((hsv_final[2]-hsv_inicial[2]) / (n_colores-1)))

  (r,g,b) = hsv_a_rgb(h,s,v)
  pintar_color(r,g,b)
```

```python
def gama_colores(color_ini, color_fin, n_color):
  t_ini = rgb_a_hsv(color_ini[0], color_ini[1], color_ini[2])
  h_ini = t_ini[0]
  s_ini = t_ini[1]
  v_ini = t_ini[2]

  t_fin = rgb_a_hsv(color_fin[0], color_fin[1], color_fin[2])
  h_fin = t_fin[0]
  s_fin = t_fin[1]
  v_fin = t_fin[2]

  lista_colores = []
  for i in range(n_colores):
    h = h_ini + (i * (h_fin - h_ini) / (n_colores - 1))
    s = s_ini + (i * (s_fin - s_ini) / (n_colores - 1))
    v = v_ini + (i * (v_fin - v_ini) / (n_colores - 1))
    t_rgb = hsv_a_rgb(h, s, v)
    lista_colores.append((t_rgb[0], t_rgb[1], t_rgb[2]))

  return lista_colores
```

```python
lista_colores = gama_colores((255,255,0), (255,0,0), 10)
print(lista_colores)
```
